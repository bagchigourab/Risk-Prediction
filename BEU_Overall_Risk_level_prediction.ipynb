```python
# Removing punctuations
def remove_punctuation(text):
    punct_str = string.punctuation
    punct_str = punct_str.replace("'", "") # discarding apostrophe from the string to keep the contractions intact
    return text.translate(str.maketrans("", "", punct_str))
```

```python
#POS
def keep_pos(text):
    tokens = regexp.tokenize(text)
    tokens_tagged = nltk.pos_tag(tokens)
    #keep_tags = ['NN', 'NNS', 'NNP', 'NNPS', 'FW']
    keep_tags = ['NN', 'NNS', 'NNP', 'NNPS', 'FW', 'PRP', 'PRPS', 'RB', 'RBR', 'RBS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WPS', 'WRB']
    keep_words = [x[0] for x in tokens_tagged if x[1] in keep_tags]
    return " ".join(keep_words)
```

```python
import pandas as pd
import numpy as np

# Assuming 'normalized text' and 'Risk Level' are the columns in your DataFrame
normalized_text = new_df['normalized text']
risk = result_df['Risk Level']

# Number of sentences you want after bootstrapping
desired_size = 1800
# Set pandas display option to show more rows
pd.set_option('display.max_rows', 100)

# Create a list to store bootstrapped samples
bootstrapped_samples = []

# Perform bootstrapping until the desired size is reached
while len(bootstrapped_samples) < desired_size:
    # Randomly sample with replacement from the original data
    bootstrap_indices_text = np.random.choice(len(normalized_text), size=len(normalized_text), replace=True)
    bootstrap_indices_risk = np.random.choice(len(risk), size=len(normalized_text), replace=True)  # Use len(risk) instead

    bootstrap_sample_text = normalized_text.iloc[bootstrap_indices_text].reset_index(drop=True)
    bootstrap_sample_risk = risk.iloc[bootstrap_indices_risk].reset_index(drop=True)

    # Append the bootstrap samples to the list
    bootstrapped_samples.append(pd.concat([bootstrap_sample_text, bootstrap_sample_risk], axis=1))

# Concatenate all the bootstrapped samples into a single dataframe
bootstrapped_data = pd.concat(bootstrapped_samples, ignore_index=True)

# Trim the excess rows if the size exceeds the desired_size
bootstrapped_data = bootstrapped_data.head(desired_size)
bootstrapped_data.columns = ['normalized text', 'Risk Level']

# Display the resulting DataFrame
print(bootstrapped_data)
```
